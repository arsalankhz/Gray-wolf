```python
import numpy as np
import math
import random

def sanitized_gwo(objective_function, lb, ub, dim, SearchAgents_no, Max_iter):

    Positions = np.zeros((SearchAgents_no, dim))

    for i in range(SearchAgents_no):

        for j in range(dim):

            Positions[i, j] = random.uniform(0, 1) * (ub[j] - lb[j]) + lb[j]

    Alpha_pos = np.zeros(dim)
    Alpha_score = float("inf")

    Beta_pos = np.zeros(dim)
    Beta_score = float("inf")

    Delta_pos = np.zeros(dim)
    Delta_score = float("inf")

    t = 0

    while t < Max_iter:

        for i in range(SearchAgents_no):

            for j in range(dim):

                Positions[i, j] = np.clip(Positions[i, j], lb[j], ub[j])


            fitness = objective_function(Positions[i, :])


            if fitness < Alpha_score:
                Alpha_score = fitness
                Alpha_pos = Positions[i, :].copy()
            elif fitness < Beta_score:
                Beta_score = fitness
                Beta_pos = Positions[i, :].copy()
            elif fitness < Delta_score:
                Delta_score = fitness
                Delta_pos = Positions[i, :].copy()

        a = 2 - t * (2 / Max_iter)

        for i in range(SearchAgents_no):
            for j in range(dim):

                r1 = random.random()
                r2 = random.random()

                C_chaotic = 0.5 + (0.5 * math.exp(-j/500)) + (1.4 * (math.exp(-0.0001 * j/Max_iter))) + ((0.5 - 2.5) * ((j/Max_iter)**2))

                A1 = 2 * a * r1 - a
                C1 = 2 * r2

                D_alpha = abs(C1 * Alpha_pos[j] - Positions[i, j])
                X1 = Alpha_pos[j] - A1 * D_alpha

                A2 = 2 * a * r1 - a
                C2 = 2 * r2
                D_beta = abs(C2 * Beta_pos[j] - Positions[i, j])
                X2 = Beta_pos[j] - A2 * D_beta

                A3 = 2 * a * r1 - a
                C3 = 2 * r2
                D_delta = abs(C3 * Delta_pos[j] - Positions[i, j])
                X3 = Delta_pos[j] - A3 * D_delta

                Positions[i, j] = (X1 + X2 + X3) / 3

        t += 1
        print(f"Iteration {t}: Best Cost = {Alpha_score}")

    return Alpha_pos, Alpha_score
```
